```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-10, 10, 400)
y = x**2 - 3*x + 2

plt.figure()
plt.plot(x, y)
plt.title("Квадратичная функция")
plt.xlabel("Ось X")
plt.ylabel("Ось Y")
plt.grid(True)
plt.show()
```

```python
x = np.linspace(0, 10, 400)
y = np.sin(x)

plt.figure()
plt.plot(x, y)
plt.title("Гармоническая функция")
plt.xlabel("Ось X")
plt.ylabel("Ось Y")
plt.grid(True)
plt.show()
```

```python
x_quad = np.linspace(-10, 10, 400)
y_quad = x_quad**2 - 3*x_quad + 2

x_harm = np.linspace(0, 10, 400)
y_harm = np.sin(x_harm)

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 6))

ax1.plot(x_quad, y_quad)
ax1.set_title("Квадратичная функция")
ax1.set_xlabel("Ось X")
ax1.set_ylabel("Ось Y")
ax1.grid(True)

ax2.plot(x_harm, y_harm)
ax2.set_title("Гармоническая функция")
ax2.set_xlabel("Ось X")
ax2.set_ylabel("Ось Y")
ax2.grid(True)

plt.show()
```

```python
plt.figure()
plt.plot(x_quad, y_quad, label='Квадратичная')
plt.plot(x_harm, y_harm, label='Гармоническая')

plt.title("Квадратичная и гармоническая функции")
plt.xlabel("Ось X")
plt.ylabel("Ось Y")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Параметрическая функция
t = np.linspace(0, 4 * np.pi, 400)
x = np.cos(3 * t)
y = np.sin(2 * t)

# Построение графика
plt.figure()
plt.plot(x, y)
plt.title("Параметрический график: x = cos(3t), y = sin(2t)")
plt.xlabel("Ось X")
plt.ylabel("Ось Y")
plt.grid(True)
plt.show()
```

```python
years = [
    2012, 2013, 2014, 2015, 2016, 2017, 
    2018, 2019, 2020, 2021, 2022
]
values = [
    4341.15, 4599.21, 4981.51, 5182.99, 4794.34, 
    4900.79, 4792.54, 4806.32, 5358.53, 6268.31, 7178.32
]

plt.figure()
plt.bar(years, values)
plt.title("Прожиточный минимум (Кыргызская Республика)")
plt.xlabel("Год")
plt.ylabel("Прожиточный минимум (в сомах)")
plt.grid(True)
plt.show()
```

```python
plt.figure()
plt.pie(values, labels=years, autopct='%1.1f%%')
plt.title("Прожиточный минимум по годам")
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D

x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)
z = np.sin(np.sqrt(x**2 + y**2))

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_wireframe(x, y, z)
ax.set_title("Каркасная поверхность")
plt.show()
```
